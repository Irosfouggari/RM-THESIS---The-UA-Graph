```python
def add_cr(abstractList, doiList):
    i = 0
    table= {"cord_uid":[], "Entity":[], "Class":[]}
    for doc in nlp_bc.pipe(abstractList):
        doi = doiList[i]
        for x in doc.ents:
          table["cord_uid"].append(doi)
          table["Entity"].append(x.text)
          table["Class"].append(x.label_)
        i +=1
    return table
```

```python
#Read in file
meta_df = pd. read_csv("C:/Users/Iro Sfoungari/Desktop/THESIS DATA/CORD-19/1000_metadata.csv")#Sort out blank abstracts
df = meta_df.dropna(subset=['abstract'])

#Create lists
doiList = df['cord_uid'].tolist()
abstractList = df['abstract'].tolist()
table = add_cr(abstractList, doiList)

trans_df = pd.DataFrame(table)
```
